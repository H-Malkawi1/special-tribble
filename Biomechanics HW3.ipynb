determining the effect of angle, speed and height to achieve the farthest distance in a vault

```python
# importing libraries
import math
from matplotlib import pyplot as plt
```

```python
#constants
g = 9.8                                             # (m/s^2) gravitational acceleration 
h = 0.55 * 1.61                                     # (0.55 times my height) 
v = 9                                               # (m/s) the initial velocity
theta = [math.radians(x) for x in range(20,71)]     # angles from 20 degrees to 70 degrees converted to radians
vary = [0.9, 1, 1.1]
```

The equations:

t_12 = 2 * (v * sin(theta) / g) equation for time 1 and 2 which is the full time of flight until y = 0 

t_3 = (v * sin(theta) +- sqrt(v ** 2 * (sin(theta)) ** 2 - 2 * h * g)) / g  is time 3; falling under y = 0

d = v * cos(theta) * ( t_12 + t_3)     where d is the horizontal range 

```python
# part 1
# observing the effect of varying h
plt.figure(1)
for i in range (3):
    d = []
    angle = 0
    first = True
    H = h * vary[i]
    end = 0
    for j in range(len(theta)):
        try:
            t3 = (v * math.sin(theta[j]) + math.sqrt(v ** 2 * (math.sin(theta[j])) ** 2 - 2 * H * g)) / g
            t12 = 2 * (v * math.sin(theta[j]) / g)
            d.append(v * math.cos(theta[j]) * (t3 + t12))
            end += 1
            if first:
                angle = j + 20
                first = False
        except ValueError:
            pass
    #plt.subplot(3, 1, i + 1)
    plt.plot(range(angle, angle + end), d)
    print(f'maximum distance = {max(d)} at angle {d.index(max(d))+angle-1}')
    print(f'range of angles = {angle, end+angle-1}')
    plt.title(f'the horizontal range when height is {round(h,3)} * {vary[i]}')
    plt.xlabel('Theta (degree)')                                                  # labeling the x-axis
    plt.ylabel('range (m)')                                                       # labeling the y-axis
    plt.xlim(25,72)
    plt.ylim(10, 16)
    plt.show()
```

```python
# part 1
# observing the effect of varying v
plt.figure(2)
for i in range (3):
    d = []
    angle = 0
    first = True
    V = v * vary[i]
    end = 0
    for j in range(len(theta)):
        try:
            t3 = (V * math.sin(theta[j]) + math.sqrt(V ** 2 * (math.sin(theta[j])) ** 2 - 2 * h * g)) / g
            t12 = 2 * (V * math.sin(theta[j]) / g)
            d.append(V * math.cos(theta[j]) * (t3 + t12))
            end += 1
            if first:
                angle = j + 20
                first = False
        except ValueError:
            pass
    #plt.subplot(3, 1, i + 1)
    plt.plot(range(angle, angle + end), d)
    print(f'maximum distance = {max(d)} at angle {d.index(max(d))+angle-1}')
    print(f'range of angles = {angle, end+angle-1}')
    plt.title(f'the horizontal range when velocity is {v} * {vary[i]}')
    plt.xlabel('Theta (degree)')                                                  # labeling the x-axis
    plt.ylabel('range (m)')                                                       # labeling the y-axis
    plt.xlim(23,75)
    plt.ylim(8, 20)
    plt.show()
```
